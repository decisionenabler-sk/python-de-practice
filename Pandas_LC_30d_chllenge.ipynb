```python
import pandas as pd

def article_views(views: pd.DataFrame) -> pd.DataFrame:
  df = views[views['author_id'] == views['viewer_id']] # only select the rows where both are equal
  df = df.drop_duplicates(subset=['author_id']) # distinct values
  df = df.sort_values(by=['author_id']) # sort by author_id
  df = df.rename(columns={'author_id':'id'}) # rename the col to id
  df = df[['id']]
  return df
data = {
    'article_id': [1, 1, 2, 2, 4, 3, 3],
    'author_id': [3, 3, 7, 7, 7, 4, 4],
    'viewer_id': [5, 6, 7, 6, 1, 4, 4],
    'view_date': ['2019-08-01', '2019-08-02', '2019-08-01', '2019-08-02', '2019-07-22', '2019-07-21', '2019-07-21']
}
article_views(pd.DataFrame(data))
```
