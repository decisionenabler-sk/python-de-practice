[Article Views I](https://leetcode.com/problems/article-views-i/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
import pandas as pd

def article_views(views: pd.DataFrame) -> pd.DataFrame:
  df = views[views['author_id'] == views['viewer_id']] # only select the rows where both are equal
  df = df.drop_duplicates(subset=['author_id']) # distinct values
  df = df.sort_values(by=['author_id']) # sort by author_id
  df = df.rename(columns={'author_id':'id'}) # rename the col to id
  df = df[['id']]
  return df
data = {
    'article_id': [1, 1, 2, 2, 4, 3, 3],
    'author_id': [3, 3, 7, 7, 7, 4, 4],
    'viewer_id': [5, 6, 7, 6, 1, 4, 4],
    'view_date': ['2019-08-01', '2019-08-02', '2019-08-01', '2019-08-02', '2019-07-22', '2019-07-21', '2019-07-21']
}
article_views(pd.DataFrame(data))
```

[Big Countries](https://leetcode.com/problems/big-countries/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
data = {
    'name': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'continent': ['Asia', 'Europe', 'Africa', 'Europe', 'Africa'],
    'area': [652230, 28748, 2381741, 468, 1246700],
    'population': [25500100, 2831741, 37100000, 78115, 20609294],
    'gdp': [20343000000, 12960000000, 188681000000, 3712000000, 100990000000]
}
world = pd.DataFrame(data)
def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    ouptut_df = world[(world['area']>=3000000) | (world['population']>=25000000)]
    ouptut_df = ouptut_df[['name', 'population', 'area']]
    return ouptut_df
big_countries(world)
```

[Recyclable and low fat products](https://leetcode.com/problems/recyclable-and-low-fat-products/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
data = [['0', 'Y', 'N'], ['1', 'Y', 'Y'], ['2', 'N', 'Y'], ['3', 'Y', 'Y'], ['4', 'N', 'N']]
products = pd.DataFrame(data, columns=['product_id', 'low_fats', 'recyclable']).astype({'product_id':'int64', 'low_fats':'category', 'recyclable':'category'})
def find_products(products: pd.DataFrame) -> pd.DataFrame:
    output_df = products[(products['low_fats'] == 'Y') & (products['recyclable'] == 'Y')]
    output_df = output_df[['product_id']]
    return output_df
find_products(products)
```

[Customer who never order](https://leetcode.com/problems/customers-who-never-order/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    merged_df = pd.merge(customers, orders, how='left', left_on='id' , right_on='customerId')
    merged_df = merged_df[merged_df['customerId'].isna()]
    merged_df = merged_df[['name']].rename(columns={'name':'Customers'})
    return merged_df
data = [[1, 'Joe'], [2, 'Henry'], [3, 'Sam'], [4, 'Max']]
customers = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
data = [[1, 3], [2, 1]]
orders = pd.DataFrame(data, columns=['id', 'customerId']).astype({'id':'Int64', 'customerId':'Int64'})
find_customers(customers,orders)
```

[Invalid tweets](https://leetcode.com/problems/invalid-tweets/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
import pandas as pd
data = [[1, 'Vote for Biden'], [2, 'Let us make America great again!']]
tweets = pd.DataFrame(data, columns=['tweet_id', 'content']).astype({'tweet_id':'Int64', 'content':'object'})

def invalid_tweets(tweets: pd.DataFrame) -> pd.DataFrame:
    output = tweets[tweets['content'].str.len() > 15]
    return output[['tweet_id']]
invalid_tweets(tweets)
```

[Ads Performance](https://leetcode.com/problems/ads-performance/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
import pandas as pd
data = [[1, 1, 'Clicked'], [2, 2, 'Clicked'], [3, 3, 'Viewed'], [5, 5, 'Ignored'], [1, 7, 'Ignored'], [2, 7, 'Viewed'], [3, 5, 'Clicked'], [1, 4, 'Viewed'], [2, 11, 'Viewed'], [1, 2, 'Clicked']]
ads = pd.DataFrame(data, columns=['ad_id', 'user_id', 'action']).astype({'ad_id':'Int64', 'user_id':'Int64', 'action':'object'})

def ads_performance(ads: pd.DataFrame) -> pd.DataFrame:
    filtered_ads = ads[ads['action'] != 'Ignored']
    count_actions = filtered_ads.groupby('ad_id')['action'].count().reset_index()
    count_actions.rename(columns={'action': 'action_count'}, inplace=True)
    count_clicks = filtered_ads[filtered_ads['action']== 'Clicked'].groupby('ad_id')['action'].count().reset_index()
    count_clicks.rename(columns={'action': 'click_count'}, inplace=True)
    ads_df = pd.merge(count_actions, count_clicks, how='left', on='ad_id')
    ads_df['ctr'] = (ads_df['click_count'] / ads_df['action_count']) * 100
    ads_df['ctr'] = ads_df['ctr'].round(2)
    ads_df = ads_df[['ad_id', 'ctr']]
    result = pd.merge(ads,ads_df, how='left', on='ad_id')
    result = result[['ad_id', 'ctr']]
    result['ctr'].fillna(0, inplace=True)
    result = result.sort_values(by=['ctr', 'ad_id'], ascending=[False, True])
    result= result.drop_duplicates(subset=['ad_id', 'ctr'])
    return result

# execute the code
ads_performance(ads)
```

[Group Sold Products By The Date](https://leetcode.com/problems/group-sold-products-by-the-date/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata)

```python
import pandas as pd
data = [['2020-05-30', 'Headphone'], ['2020-06-01', 'Pencil'], ['2020-06-02', 'Mask'], ['2020-05-30', 'Basketball'], ['2020-06-01', 'Bible'], ['2020-06-02', 'Mask'], ['2020-05-30', 'T-Shirt']]
activities = pd.DataFrame(data, columns=['sell_date', 'product']).astype({'sell_date':'datetime64[ns]', 'product':'object'})
def categorize_products(activities: pd.DataFrame) -> pd.DataFrame:
    grouped_df = activities.groupby('sell_date')
    output = grouped_df.agg(
      num_sold=('product', 'nunique'),
      products=('product', lambda x: ','.join(sorted(set(x))))
    ).reset_index()
    output.sort_values('sell_date', inplace=True)
    return output
categorize_products(activities)
```
